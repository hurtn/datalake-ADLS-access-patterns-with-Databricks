```
# read data in delta format using direct path
loansdf = spark.read\
  .format("delta")\
  .load("abfss://test-container@rrdemostorageacc.dfs.core.windows.net/loans")

display(loansdf)
```

```
iot_devices = spark.read\
  .format("delta")\
  .load("abfss://test-container@rrdemostorageacc.dfs.core.windows.net/iot-devices")

display(iot_devices)
```

```
loansdf = spark.read.table("testdb.loans")

display(loansdf)
```
